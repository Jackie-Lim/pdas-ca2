## "Which industry offers the best overall job stability and prospects for fresh graduates in Singapore, based on employment outcomes and salary consistency"

Job stability means how likely a person is able to keep their job.

A job with high stability usually have:
- high employment rate
- high number of years worked in a company
- stable salary accross the years



```python
import pandas as pd

df1 = pd.read_csv("graduate_employment_1.csv")
df2 = pd.read_csv("graduate_employment_2.csv")

df = pd.merge(df1, df2, on="Key", how="inner")

# Drop columns with all NaN rows
df = df.dropna(axis=1, how="all")

# Drop columns with duplicated content
df = df.drop(columns=df.columns[df.T.duplicated()])

# Drop rows with ANY NaN column
df = df.dropna()

df.head()
```
